```scala
// Define GCS paths
val bucketName = "movielens_dataset"
val moviesPath = s"gs://$bucketName/movie.csv"
```

```scala
import org.apache.spark.sql.SparkSession

val spark = SparkSession.builder()
  .appName("Avro Format")
  .getOrCreate()

val moviesDF = spark.read
  .format("csv")
  .option("header", "true")
  .option("inferSchema", "true")
  .load(moviesPath)
```

```scala
println(s"movies: ${moviesDF.count()}")
```

```scala
// Remove duplicates based on 'movieId' and 'title' columns
val cleanedMoviesDF = moviesDF.dropDuplicates(Seq("movieId", "title"))
```

```scala
println(s"movies: ${cleanedMoviesDF.count()}")
```

```scala
%AddDeps org.apache.spark spark-avro_2.12 3.3.2 --transitive
```

```scala

// Save the cleaned DataFrame as Avro in GCP Cloud Storage
cleanedMoviesDF.write.format("avro").save("gs://movielens_dataset/cleaned-movies")
```

```scala
spark.stop()
```

```scala
// Define GCS paths
val bucketName = "movielens_dataset"
val ratingsPath = s"gs://$bucketName/rating.csv"
```

```scala
import org.apache.spark.sql.SparkSession

val spark = SparkSession.builder()
  .appName("Year Wise Saving")
  .getOrCreate()

val ratingsDF = spark.read
  .format("csv")
  .option("header", "true")
  .option("inferSchema", "true")
  .load(ratingsPath)
```

```scala
ratingsDF.show()
```

```scala
import org.apache.spark.sql.functions._

val ratingsWithYearDF = ratingsDF.withColumn(
  "year",
  year(to_timestamp(col("timestamp"), "yyyy-MM-dd HH:mm:ss"))
)
```

```scala
ratingsWithYearDF.write
  .partitionBy("year")
  .parquet("hdfs:///user/caseStudy5/")
```

```scala
spark.stop()
```
